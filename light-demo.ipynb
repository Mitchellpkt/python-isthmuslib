## Import Library

```python
import isthmuslib as isli
isli.version()
```

### Get sample data set (#1)

```python
x,y = isli.demoData(1)
```

### Histogram
This is a wrapper for pyplot hist (1d) as a one-liner

```python
isli.hist(y, xlabel='Eigenvalues', title='Decomposition');
```

### Scatterplot
This is a wrapper for pyplot scatter as a one-liner

```python
isli.scatter(x,y, markersize=250, xlabel='resonance (w)', ylabel='frequency (f)', markercolor='blue');
```

### Heatmap (hist2d)
This is a wrapper for pyplot hist2d as a one-liner with some better log bin logic

```python
isli.hist2d(x,y, xlabel='resonance (w)', ylabel='frequency (f)', cmap='jet', bins=[5,10]);
```

### Dictionary scatter
This is a wrapper for pyplot but has a different input style. Each data set is flexibly described in a dictionary, and multiple data sets can be loaded as a list of dictionaries

dataDict = isli.demoData(2)

```python
dataDict = isli.demoData(2)
dataDict[1]['markersize']=400 # You can edit parameters in place
dataDict
```

```python
isli.scatterDictionary(dataDict, xlabel='time', ylabel='entropy', legend=True, lines=True, linestyle='--');
```
