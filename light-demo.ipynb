## Import Library

```python
import isthmuslib as isli
```

### Get sample data set (#1)

```python
x,y = isli.demoData(5)
```

### Histogram
This is a wrapper for pyplot hist (1d) as a one-liner

```python
isli.hist(y, xlabel='Eigenvalues', title='Decomposition', bins=40);
```

### Histogram
the isthmuslib hist function supports multiple sets in `{"label":data}` format

```python
dataDict = {"Neutral (pH=6.93)":x, "Acidic (pH=4.47)":y}
isli.hist(dataDict, xlabel='Wavelength', title='Spectra', alpha=0.5, bins=40);
```

### Scatterplot
This is a wrapper for pyplot scatter as a one-liner

```python
isli.scatter(x,y, watermarkText='test watermark text', markersize=250, xlabel='resonance (w)', ylabel='frequency (f)', markercolor='blue');
```

### 2D Histogram
This is a wrapper for pyplot hist2d as a one-liner with some better log bin logic

```python
isli.hist2d(x,y, xlabel='resonance (w)', ylabel='frequency (f)', cmap='jet', bins=[5,10]);
```

### Dictionary scatter
This is a wrapper for pyplot but has a different input style. Each data set is flexibly described in a dictionary, and multiple data sets can be loaded as a list of dictionaries

dataDict = isli.demoData(2)

```python
dataDict = isli.demoData(2)
dataDict[1]['markersize']=400 # You can edit parameters in place
dataDict
```

```python
isli.scatterDictionary(dataDict, xlabel='time', ylabel='entropy', legend=True, lines=True, linestyle='--');
```

### Plotting data frames

```python
import pandas as pd
df = pd.DataFrame()
df['x'] = dataDict[0]['x']
df['y1'] = dataDict[0]['y']
df['y2'] = dataDict[1]['y']
df.head()
```

```python
isli.shotgunPlots1d(df, useFields=['y1','y2'])
```

```python
isli.shotgunPlots2d(df, xColName='x')
```

### Heatmaps

```python
df = isli.demoData(3)
isli.heatmapDataFrame(df,xcol='x',ycol='y', zcol='z', figsize=(8,5));
```

```python
isli.heatmap(df['x'],df['y'],df['z'], vmax=1000, cmap="YlGnBu", linecolor='k', figsize=(8,5), title='Heatness', xlabel='time', ylabel='height');
```
